```python
height = np.random.normal(160, 30, 300)
weight = 0.3*height+np.random.normal(30, 10, 300)


w_hat = height * 0.3 + 30
```

```python
plt.rcParams['xtick.color'] = 'white'
plt.rcParams['ytick.color'] = 'white'
```

```python
plt.rcParams['xtick.color'] = 'w'
plt.rcParams['ytick.color'] = 'w'
```

```python
fig = plt.figure(figsize=(6, 3), dpi=120, facecolor='black')
ax = fig.subplots() # facecolor

ax2 = fig.add_axes([0.2, 0.6, 0.3, 0.2])
# [x, y]
ax2.set_facecolor('k')
a = np.linspace(-5, 5, 100)
ax2.plot(a, a**2)
ax2.plot(a, a**3)
ax2.plot(a, a)


ax.set_facecolor('k')

ms = ax.scatter(height, weight, marker='o',  facecolor='w', edgecolor='k', s=50, label='weight') # fc, ec
ls = ax.plot(height, w_hat, color='r', linewidth=3, linestyle=':', label='w_hat') # lw, ls

ax.set_xlabel("Height", color='w', fontsize=20)
ax.set_ylabel("Weight", color='w', fontsize=20)

ax.set_title("Basics of Figure and Axes", color='red')
ax.set_xlabel("height")
ax.set_ylabel("weight")

ticks = ax.set_xticks(np.arange(80, 241, 10))
ax.set_xticklabels(ax.get_xticks(), fontsize=5, rotation=45)
ax.legend(loc='lower right')
ax.grid()
plt.show()
```
