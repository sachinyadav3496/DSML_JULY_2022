```python
rad = np.linspace(0, 10*np.pi, 100)
sin = np.sin(rad)
cos = np.cos(rad)
fig = plt.figure(figsize=(12,5), dpi=120)

ax1 = fig.add_axes([0, 0, 1, 1])
#ax2 = fig.add_axes([0.7, 0.7, 0.2, 0.2])

ax1.plot(rad, sin, "--c*", rad, cos, "-.y8", label="$Sin\\theta$", lw=2, ms=10, mfc='w')


ax1.spines["bottom"].set_color("c")
ax1.spines["left"].set_color("y")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)

ax1.set_xlim(-1, 38)
ax1.set_ylim(-1.2, 1.2)

ax1.set_xticks(np.arange(0, 35, 2))
ax1.set_xticklabels([chr(65+i) for i in range(18)], fontsize=15, color="c")

ax1.legend(fontsize=20, loc="upper right")
plt.show()
```

```python
rad = np.linspace(0, 10*np.pi, 100)
sin = np.sin(rad)
cos = np.cos(rad)
fig = plt.figure(figsize=(15,7), dpi=120)

ax1 = fig.add_axes([0, 0, 1, 1])
#ax2 = fig.add_axes([0.7, 0.7, 0.2, 0.2])
ax1.plot(rad, sin, "--c*", rad, cos, "-.y8", label="$Sin\\theta$", lw=4, ms=20, mfc='w')
# lw line width, ms marker size, mfc marker face color
ax1.spines["bottom"].set_color("c")
ax1.spines["left"].set_color("y")
ax1.spines["top"].set_visible(False)
ax1.spines["right"].set_visible(False)
#ax1.set_xlim(-1, 38)
#ax1.set_ylim(-1.2, 1.2)

angle = np.arange(0, 1800, 30)
ax1.set_xticks(np.linspace(0, 10*np.pi, 60))
ax1.set_xticklabels(angle, fontsize=15, color="c", rotation=90)

ax1.set_yticks(np.round(np.arange(-1, 1.1, .1),2))
ax1.set_yticklabels(ax1.get_yticks(), fontsize=15, color='y')

ax1.grid(ls='--', color='r', alpha=0.2)
ax1.legend(fontsize=20, loc="upper right")

ax1.set_xlabel("Angles", color="cyan", fontsize=30)
ax1.set_ylabel("sin/cos", color="red", fontsize=30)

ax1.set_title("Sin vs Cos", fontsize=40, color="salmon", pad=20)

plt.tight_layout()
plt.savefig("sin_vs_cos.jpg")
plt.show()
```

```python
data = sns.load_dataset("flights")
```

```python
data.head()
```

```python
tmp = data.groupby("year").agg({"passengers": "sum"}).reset_index()
tmp.head()
```

```python
tmp2 = pd.DataFrame({
    "year": tmp['year'],
    "passengers": tmp.passengers+np.random.randint(-500, 500, 12)
})
tmp2
```

```python
plt.figure(figsize=(8, 4), dpi=200)
plt.plot("year", "passengers", "--oc", data=tmp, mfc='w', label="air india")
plt.plot("year", "passengers", "--sm", data=tmp2, mfc='w', label="air asia")
plt.grid()
ax = plt.gca()
for sp in ['top', 'right']:
    ax.spines[sp].set_visible(False)
ax.set_title("Passengers by Year")

ax.set_yticks(np.arange(1000, 5500, 500))
ax.set_xticks(tmp.year)
plt.legend()
plt.show()
```
