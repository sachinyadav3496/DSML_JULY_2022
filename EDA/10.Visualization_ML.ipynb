```python
obs = plt.boxplot(tips['total_bill'], patch_artist=True)
obs["boxes"][0].set_lw(2)
obs["boxes"][0].set_ls('--')
obs["boxes"][0].set_facecolor('yellow')
obs['whiskers'][0].set_color('red')
obs['fliers'][0].set_mfc('red')
obs['fliers'][0].set_mec('white')
obs['fliers'][0].set_ms(10)

```

```python
fig, ax = plt.subplots(figsize=(6, 4), dpi=120)
ax = tips['sex'].value_counts().plot(kind='bar', ax=ax)
for p,c in zip(ax.patches, ['red', 'green']):
    p.set_fc(c)
    x = p.get_x()+0.1
    t = p.get_height()
    y = t // 2
    ax.text(x, y, t, color='white', fontsize=20)

for sp in ax.spines:
    ax.spines[sp].set_visible(False)
    
ax.set_yticks([])
ax.set_title('Male vs Female')
plt.show()
```

```python
plt.figure(dpi=120)
tips['day'].value_counts().plot(kind='pie', counterclock=False,
                               startangle=90, shadow=True,
                               explode=[0.05, 0.05, 0.05, 0.2],
                               autopct="%d%%",
                               colors=["grey", "grey", "grey", "red"],
                               pctdistance=1.2,
                               labeldistance=0.7,
                           )
cr = mpl.patches.Circle((0,0), 0.5, fc='w')
ax = plt.gca()
ax.add_artist(cr)
ax.set_ylabel("")
plt.tight_layout()
plt.show()
```

### Heat Map

    2D Matrix

```python
flights = sns.load_dataset('flights')
```

```python
flights.head()
```

```python
df = flights.pivot_table(index='year', columns='month', values='passengers')
```

```python
df
```

```python
plt.figure(dpi=120)
plt.matshow(df, cmap="coolwarm", fignum=1)
plt.colorbar()
plt.xticks(np.arange(len(df.columns)), df.columns, rotation=90)
plt.yticks(np.arange(df.shape[0]), df.index)
plt.show()
```

```python
plt.figure(dpi=120)
plt.matshow(df, cmap="binary", fignum=1)
plt.colorbar()
plt.xticks(np.arange(len(df.columns)), df.columns, rotation=90)
plt.yticks(np.arange(df.shape[0]), df.index)
plt.show()
```

```python
fig, ax = plt.subplots(dpi=120)
sns.heatmap(df, cmap='binary', ax=ax)
plt.show()
```

```python
sns.clustermap(df, cmap='coolwarm')
plt.show()
```
