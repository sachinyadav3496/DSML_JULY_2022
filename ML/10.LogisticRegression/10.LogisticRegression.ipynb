```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.linspace(-5, 5, 100)
```

###### Sigmoid


$$ sigmoid = \frac {1} {1+e^{-t}}$$

```python
def sigmoid(t):
    return (1/(1+np.exp(-t)))
```

```python
y = sigmoid(t)
```

```python
plt.figure(dpi=100, figsize=(6, 2))
plt.plot(t, y)
plt.xlabel("t")
plt.ylabel("sigmoid")
plt.show()
```

```python
h1 = (np.random.random(50)*5).reshape(-1, 1)
h2 = (5+ np.random.random(50)*7).reshape(-1, 1)
X = np.r_[h1, h2] # Study Hours
pass_ = np.ones(50)
fail_ = np.zeros(50)
y = np.append(fail_, pass_) # Result
```

```python
plt.figure(dpi=120, figsize=(6,3))
plt.plot(X, y, 'bx')
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.yticks([0, 1], ["Fail(0)", "Pass(1)"])
plt.xticks(np.arange(0, 14, 0.5), rotation=90)
plt.show()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X, y)
```

```python
y_hat = model.predict(X)
```

```python
plt.figure(dpi=120, figsize=(6,3))
plt.plot(X, y, 'bx', label="actual")
plt.xlabel("Study Hours")
plt.ylabel("Result")
plt.yticks([0, 0.5, 1], ["Fail(0)", "0.5", "Pass(1)"])
plt.xticks(np.arange(0, 14, 0.5), rotation=90)
plt.plot(X, y_hat, 'r', label='predict')
plt.hlines(0.5, 0, 14, ls='--', color='k', label="Decision Boundary")
plt.legend()
plt.show()
```

```python
def predict(y_hat, threshhold=0.5):
    y = np.zeros(len(y_hat))
    y[y_hat >= threshhold] = 1
    return y
```

```python
y_hat = predict(model.predict(X))
```
