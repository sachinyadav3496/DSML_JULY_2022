```python
x = 6*np.random.random(100)-3
```

```python
y = 0.3*x**2+x+np.random.random(100)*2
```

```python
plt.scatter(x, y)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
lin_model = LinearRegression()
lin_model.fit(x.reshape(-1, 1), y)
```

```python
y_lin = lin_model.predict(x.reshape(-1, 1))
plt.scatter(x, y)
plt.scatter(x, y_lin)
plt.xlabel("X")
plt.ylabel("y")
plt.show()
```

```python
plt.scatter(x, y)
plt.scatter(x, y_lin)
plt.xlabel("X")
plt.ylabel("y")
```

```python
plt.scatter(x**2, y)
plt.scatter(x**2, LinearRegression().fit((x**2).reshape(-1, 1), y).predict((x**2).reshape(-1, 1)))
plt.xlabel("X")
plt.ylabel("y")
```

```python
X_tr = np.c_[x.reshape(-1, 1), (x**2).reshape(-1, 1)]
```

```python
m = LinearRegression()
m.fit(X_tr, y)
```

```python
y_poly = m.predict(X_tr)
```

```python
plt.figure(dpi=120)
plt.scatter(x, y, fc="w", s=200, ec='b')
plt.plot(x, y_lin, color='k', lw=2)
plt.scatter(x, y_poly, color='red', fc="w", s=50)
plt.xlabel("X")
plt.ylabel("y")
plt.tight_layout()
plt.show()
```

```python
X_tr1 = np.c_[x.reshape(-1, 1), (x**2).reshape(-1, 1), (x**3).reshape(-1, 1)]
```

```python
model = LinearRegression()
model.fit(X_tr1, y)
y_3 = model.predict(X_tr1)
```

```python
plt.figure(dpi=120)
plt.scatter(x, y, fc="w", s=200, ec='b')
plt.plot(x, y_lin, color='k', lw=2)
plt.scatter(x, y_poly, color='red', fc="w", s=50)
plt.xlabel("X")
plt.ylabel("y")
plt.scatter(x, y_3, color='cyan', fc="w", s=50)
plt.tight_layout()

plt.show()
```
